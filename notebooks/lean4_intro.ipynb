## 3. 簡単な定理証明

`theorem` キーワードで定理を定義し、`rfl` や `simp` などのタクティクで証明します。

```lean4
-- 自然数の加算に関する基本定理: n + 0 = n
theorem my_add_zero (n : Nat) : n + 0 = n := by
  rfl
```

```lean4
-- 加算の交換法則
theorem my_add_comm (a b : Nat) : a + b = b + a := by
  omega
```

```lean4
-- 証明した定理を確認
#check my_add_zero
#check my_add_comm
```

## 4. lean4_viz を使った数式可視化

`lean4_viz` モジュールを使って、LaTeX形式で数式をレンダリングします。

```lean4
import lean4_viz
```

```lean4
# 自然数の加算定理をLaTeXで表示
lean4_viz.show_latex(r"\forall n \in \mathbb{N},\ n + 0 = n")
```

```lean4
# 加算の交換法則
lean4_viz.show_latex(r"\forall a, b \in \mathbb{N},\ a + b = b + a")
```

```lean4
# Lean4の識別子をテキスト形式で表示
lean4_viz.show_expr("Nat.add_zero : ∀ (n : Nat), n + 0 = n")
```

```lean4
# 複雑な数式の例
lean4_viz.show_latex(r"\sum_{k=0}^{n} k = \frac{n(n+1)}{2}")
```
